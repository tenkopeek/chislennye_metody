## Задание 1. Графическое отделение корней

В практической работе №2 мы уже находили интервалы с корнями, поэтому для последующей работы возьмём корень на интервале $[0, 2.5]$

## Задание 2. Метод хорд

В методе хорд неподвижным концом выбирается тот, где знак функции совпадает со знаком второй производной $f(x) \cdot f''(x) > 0$.
Для $f(x) = 10\cos(x) - 0.1x^2$:
* $f'(x) = -10\sin(x) - 0.2x$
* $f''(x) = -10\cos(x) - 0.2$

```python
def f(x):
    return 10 * np.cos(x) - 0.1 * x**2

# Параметры
a, b = 0.0, 2.5
eps = 0.001
data_chords = []

# Анализ для выбора неподвижной точки:
# f(0) = 10 (положительное значение)
# f(2.5) ≈ -8.635 (отрицательное значение)
# На этом интервале функция убывает. 
# Выберем точку b = 2.5 в качестве неподвижной для расчета.

x_n = a
iteration = 0

while True:
    # Формула метода хорд с неподвижной точкой b
    x_next = x_n - (f(x_n) * (b - x_n)) / (f(b) - f(x_n))
    diff = abs(x_next - x_n)
    
    data_chords.append([iteration + 1, x_n, x_next, f(x_next), diff])
    
    if diff < eps:
        break
    x_n = x_next
    iteration += 1

# Вывод таблицы
df_chords = pd.DataFrame(data_chords, columns=["Итерация", "x_n", "x_n+1", "f(x_n+1)", "|x_n+1 - x_n|"])
display(df_chords)

print(f"Уточненное значение корня: {x_next:.4f}")
```

## Задание 3. Метод касательных (Ньютона)

```python
def f(x):
    return 10 * np.cos(x) - 0.1 * x**2

def df(x):
    # Производная функции: f'(x) = -10*sin(x) - 0.2*x
    return -10 * np.sin(x) - 0.2 * x

# Параметры для интервала [0, 2.5]
x_n = 2.5  # Выбираем правую границу в качестве начального приближения
eps = 0.001
data_newton = []
iteration = 0

while True:
    f_xn = f(x_n)
    df_xn = df(x_n)
    
    # Вычисляем следующее приближение
    x_next = x_n - f_xn / df_xn
    diff = abs(x_next - x_n)
    
    # Сохраняем данные для таблицы
    data_newton.append([iteration + 1, x_n, x_next, round(f(x_next), 6), diff])
    
    # Проверка на достижение точности
    if diff < eps:
        break
        
    x_n = x_next
    iteration += 1

# Оформление вывода
df_newton = pd.DataFrame(data_newton, columns=["Итерация", "x_n", "x_n+1", "f(x_n+1)", "Уточнение (|x_n+1 - x_n|)"])

display(df_newton)

print(f"Итоговый корень: {x_next:.3f}")
print(f"Проверка: f({x_next:.3f}) = {f(x_next):.6f}")
```
