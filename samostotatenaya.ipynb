Уравнение: $\ln(x+5)=\cos(x)$. В форме $f(x) = 0$:

$\ln(x+5)-cos(x)=0$

```python
from numpy import log, sin, cos

def f(x):
    return log(x + 5) - cos(x)
```

## Задание 4.

Вычислите один корень заданного уравнения с помощью программы для компьютера c точностью $10^{-6}$, используя комбинированный метод хорд и касательных. 

```python
def df(x):
    # Производная: 1/(x+5) + sin(x)
    return 1 / (x + 5) + sin(x)

def combined_method(a, b, eps):
    # Проверка условия сходимости: f(a)*f(b) < 0
    if f(a) * f(b) > 0:
        print("На данном отрезке корней может не быть или их четное количество.")
        return None

    iteration = 0
    while abs(b - a) > eps:
        iteration += 1
        
        # Метод хорд для одного конца
        # Выбираем, какой конец двигать хордой, а какой касательной
        # (Условие f(x)*f''(x) > 0 определяет выбор для метода Ньютона)
        
        # Упрощенная схема комбинированного метода:
        a_new = a - (f(a) * (b - a)) / (f(b) - f(a))
        b_new = b - f(b) / df(b)
        
        a, b = a_new, b_new
        
    return (a + b) / 2, iteration

# Параметры поиска
start_a = -4.5
start_b = -4.0
accuracy = 1e-6

root, iterations = combined_method(start_a, start_b, accuracy)

if root:
    print(f"Корень уравнения: {root:.6f}")
    print(f"Количество итераций: {iterations}")
```
