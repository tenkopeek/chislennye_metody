## Задание 1. Метод итераций

Для итераций используем формулу: $x_{n+1} = \arccos(0.01 \cdot x_n^2)$.

```python
# Определяем функцию phi(x) из уравнения x = phi(x) (x = arccos(0.01*x^2))
def phi(x):
    return np.arccos(0.01 * x**2)

# Параметры метода
x0 = 1.5  # начальное приближение из графического метода
eps = 0.001
data_iter = []

x_curr = x0
iteration = 0
converged = False

# Итерационный процесс
while not converged:
    x_next = phi(x_curr)
    delta = abs(x_next - x_curr)
    
    # Записываем данные для расчетной таблицы (как в калькуляторе)
    data_iter.append([iteration + 1, round(x_curr, 5), round(x_next, 5), round(delta, 5)])
    
    if delta < eps:
        converged = True
    else:
        x_curr = x_next
        iteration += 1
        # if iteration > 100:
        #     break

# Оформление таблицы
columns = ["Итерация", "x_n", "x_n+1", "|x_n+1 - x_n|"]
df_iter = pd.DataFrame(data_iter, columns=columns)

display(df_iter)

print(f"Найденный корень: {round(data_iter[-1][2], 3)}")
print(f"Точность достигнута за {len(data_iter)} итераций.")
```
